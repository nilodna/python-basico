```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# carregar arquivo com a lista de especies
lista_especies = open('../lista_de_especies.txt', 'r').read()
lista_especies = lista_especies.split('\n')

# pegamos as ocorrências únicas, ou seja, o nome de cada espécie que aparece na lista
todas_especies = set(lista_especies)

# criamos um dicionario vazio que será preenchido mais pra frente
dicionario_especies = {}

# iteramos pelas especies
for especie in todas_especies:
    # podemos obter a ocorrência de uma variavel em um lista usando o método count()
    ocorrencias = lista_especies.count(especie)
    
    # adicionamos um novo registro ao dicionário
    dicionario_especies[especie] = ocorrencias

print(dicionario_especies)
```

```python
especies = []
ocorrencias = []

for key,value in dicionario_especies.items():
    especies.append(key)
    ocorrencias.append(value)

# criar figura
plt.figure(figsize=(15,5))
# plotar uma linha onde especies é x e ocorrencias é y
plt.plot(especies, ocorrencias)
# plotar marcadadores
plt.scatter(especies, ocorrencias)
# gerar uma grade ao fundo da figura
plt.grid('--', alpha=.3)
# inserir um nome para o eixo y
plt.ylabel('Ocorrências')
# rotacionar os nomes das especies no eixo x
plt.tick_params('x', labelrotation=45)
```
