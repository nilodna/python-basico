A planilha excel que vamos ler contém dados de maré.

As abas também contém gráficos e informações extras como amplitude, marés máxima e minima, nível do mar e médias.

Para complicar um pouco mais algmas células da planilha estão mescladas, o que dificulta ainda mais a leitura no python.

A primeira coisa que precisamos fazer é importar as bibliotecas que vão ser usadas

Para limpar a tabela e ficar mais facil de trabalhar vamos remover algumas linhas e colunas que não são necessárias, para o nosso objetivo.

````python
#Lendo o arquivo ja seleceionando a aba do mês, definindo a linha com nome das colunas, excluindo linhas desnecessarias e e definindo coluna dos index
df = pd.read_excel(k, sheet_name=mes[1], header=0, skiprows=[0,1,2,3,4], index_col=0)

#Removendo colunas desnecessárias
#Removendo coluna da hora 00:00 que repete o dado das 24:00
df.drop(df.columns[0], axis=1, inplace=True)
#Removendo colunas das maŕes minimas e máximas, aplitude e nível do mar
df.drop(df.columns[24:], axis=1, inplace=True)

#Para organizar os dias temos que excluir as linhas que vão além da quantidade de dia de cada mês
#Usando o pandas.Period para conferir quando dias tem no determinado mês do determinado ano
total_dias = pd.Period(k[-8:-4]+'-'+str(2)).days_in_month
#Removendo todas as linhas que a mais a quantidade de dia do mês em questão
df.drop(df.index[total_dias:], inplace=True)
df
````

Beleza, a gente ja consegue lipar a tabela. Mas os dados ainda estão separados em dia e hora no formato de matriz 2D. Nós queremos um série temporal da maré.
Para isso escolhi transformar esse dados em uma lista contendo os dados de todos os meses de todos os anos. A idéia é pegar cada coluna e ir "colando" embaixo da outra. 

````python
#Criando uma lista em branco para juntar todos os dados de maré
listnm = list()

#Loop para transformar a matrix de dia e horas de maré em uma lista de dados corridos
for r in range(df.shape[0]):
    row = df.iloc[r]
#     print(row)

    listnm.extend(row)
listnm
````

Agora vamos fazer isso dentro daquele loop para ler todos os anos e mes de uma vez só.

````python
#Criando uma lista em branco para juntar todos os dados de maré
listnm = list()

#Loop para ler todos os arquivos .xls arquivo por arquivo
for k in name:

    print(k)
    
    c=1 #contador de mês
    #Loop para acessar cada aba de mês dentro do arquivo
    for i in mes:

        print(i)
        
        #Lendo o arquivo ja seleceionando a aba do mês, definindo a linha com nome das colunas, excluindo linhas desnecessarias e e definindo coluna dos index
        df = pd.read_excel(k, sheet_name=i, header=0, skiprows=[0,1,2,3,4], index_col=0)

        #Removendo colunas desnecessárias
        #Removendo coluna da hora 00:00 que repete o dado das 24:00
        df.drop(df.columns[0], axis=1, inplace=True)
        #Removendo colunas das maŕes minimas e máximas, aplitude e nível do mar
        df.drop(df.columns[24:], axis=1, inplace=True)

        #Para organizar os dias temos que excluir as linhas que vão além da quantidade de dia de cada mês
        #Usando o pandas.Period para conferir quando dias tem no determinado mês do determinado ano
        total_dias = pd.Period(k[-8:-4]+'-'+str(c)).days_in_month
        #Removendo todas as linhas que a mais a quantidade de dia do mês em questão
        df.drop(df.index[total_dias:], inplace=True)
        

        c += 1
        
        #Loop para transformar a matrix de dia e horas de maré em uma lista de dados corridos
        for r in range(df.shape[0]):
            row = df.iloc[r]
#             print(row)
            listnm.extend(row)
# listnm

len(listnm)
````

````python
#Criando um vetor com as datas de cada dados de onda
dateRange = pd.date_range(start='01-Jan-%s 01:00'%(int(name[0][-8:-4])), end='01-Jan-%s 00:00'%(int(name[-1][-8:-4])+1), freq='1H')

print(len(dateRange))
dateRange
````

````python
#Criando um novo dataframe da série temporal com a variavel nível do mar e como index a data
df_final = pd.DataFrame(listnm, columns=['Nível do Mar'], index = dateRange)
df_final
````

A série temporal esta pronta!

````python
#Importanto as bibliotecas a serem usadas
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import matplotlib.pyplot as plt


#caminho para os arquivo de dados de nivél do mar
path = '../dados/'
````

````python
#Lendo o arquivo .csv da série temporal
df = pd.read_csv(path+'mare_exemplo.csv', header=0, index_col=0)
# df

# # visualizacao simples
df.plot(figsize=(25,5))
plt.title('Série temporal (fake) do nível do mar bruto', fontsize=20)
_= plt.ylabel('Nível do mar (m)', fontsize=14)
````

````python
#Transformando o index em um datetime
df.index = pd.to_datetime(df.index)
df.index
````

````python
#Slice temporal
_= df['1994-01': '1994-01'].plot()
````

````python
#Slice temporal para selecionar um período (1994)
df1994 = df['1994-03': '1994-09'].copy(deep=True)
print(df1994.shape)
````

````python
#Detecção de outlier por desvio padrão (controle de qualidade dos dados)
avg = df1994['Nível do Mar'].mean()
std = df1994['Nível do Mar'].std()

limite_inferior = avg - 3*std
limite_superior = avg + 3*std

limite_inferior, limite_superior
````

````python
ax = df1994['Nível do Mar'].plot(figsize=(25,5))
_= ax.axhline(limite_inferior, color='k')
_= ax.axhline(limite_superior, color='k')
````

````python
#Criando um filtro simples para excluir tudo que esta acima ou abaixo da média +- 3 desvio padrão
filtro_simples = ((df1994['Nível do Mar'] > limite_inferior ) & (df1994['Nível do Mar'] < limite_superior))
filtro_simples
````

````python
#Aplicando filtro na série temporal
df1994 = df1994[filtro_simples]
print(df1994.shape)
_= df1994['Nível do Mar'].plot()
````

````python
#Retirando a média da série para que ela varie em torno do zero
df1994['serie_nova'] = df1994 - df1994.mean()

#plotando as duas séries juntas
_ = df1994.plot(figsize=(25,8))
````
